```python
import pandas as pd
from math import sqrt
from sklearn.metrics import confusion_matrix 
from sklearn.metrics import accuracy_score 
from sklearn.metrics import classification_report 

#importing the dataset
df = pd.read_csv("NBA_Data/nba_logreg.csv")
```

```python
#Treating null values
print(df.isna().sum())
df['3P%'] = df['3P%'].fillna((df['3P%'].mean()))
print(df.isna().sum())
```

```python
df.head()
```

```python
#Dividing the dataset to test and train
train = df[0:1000]
test = df[1000:]
del train['Name'] 
del test['Name']

t_test = test['TARGET_5Yrs'].tolist()

k = 51
```

```python
def euclidean_distance(row1, row2):
    distance = 0.0
    for i in range(len(row1)-1):
        distance += (row1[i] - row2[i])**2
    return sqrt(distance)
```

```python
def sorting_vals(distances,k):
    distances.sort(key = lambda x: x[0])
    distances = distances[:k]
    list_zero = list(filter(lambda x: (x[1]%2 == 0) , distances))
    leng_va = len(list_zero)
    m = (k/2)+1
    if(leng_va >m):
        return 0
    else:
        return 1
```

```python
distances = list()
target = list()
       
for test_row in test.iterrows():
    test_row = list(test_row)
    test_row = test_row[1]
    #clear lists
    distances.clear()
    for train_row in train.iterrows():
        train_row = list(train_row)
        train_row = train_row[1]   
        dist = euclidean_distance(test_row, train_row)
        #print(dist)
        distances.append([dist, train_row[-1]])
    target.append(sorting_vals(distances,k))    
print(target)
```

```python
conf_mat = confusion_matrix(t_test, target) 
print("Confusion Matrix")
print(conf_mat)
    
s = conf_mat[0][0]+conf_mat[1][1]
accuracy = (s/340)*100
print("Accuracy")
print(accuracy)
print("Report :")
print(classification_report(t_test, target))
```
