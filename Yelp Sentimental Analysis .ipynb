```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
r = requests.get('https://www.yelp.com/biz/carmelinas-boston-2')
```

```python
r.status_code
```

```python
# r.text
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
answer=soup.findAll('span',class_="lemon--span__373c0__3997G raw__373c0__3rKqk")
```

```python
# answer.text
```

```python
review = []
for p in answer:
    review.append(p.text)
```

```python
review
```

```python
#Analysis the Data 
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(np.array(review), columns=['reviews'])
```

```python
#df.head()
```

```python
len(df['reviews'])
```

```python
df['word_count'] = df['reviews'].apply(lambda x: len(x.split()))
```

```python
#df.head
```

```python
df['char_count'] = df['reviews'].apply(lambda x: len(x))
```

```python
def average_words(x):
    words = x.split()
    return sum(len(word) for word in words) / len(words)
```

```python
df['average_word_length'] = df['reviews'].apply(lambda x: average_words(x))
```

```python
#counting the Stop words
```

```python
from nltk.corpus import stopwords
```

```python
stop_words = stopwords.words('english')
```

```python
len(stop_words)
```

```python
df['stopword_count'] = df['reviews'].apply(lambda x: len([word for word in x.split() if word.lower() in stop_words]))
```

```python
df.head()
```

```python
df['stopword_rate'] =  df['stopword_count'] / df['word_count']
```

```python
#df.head()
```

```python
df.sort_values(by='stopword_rate')
```

```python
df.describe()
```

```python
#Data Cleaning 
```

```python
df['lowercase'] = df['reviews'].apply(lambda x: " ".join(word.lower() for word in x.split()))
```

```python
df.head()
```

```python
df['punctuation'] = df['lowercase'].str.replace('[^\w\s]', '')
```

```python
from nltk.corpus import stopwords
```

```python
stop_words = stopwords.words('english')
```

```python
df['stopwords'] = df['punctuation'].apply(lambda x: " ".join(word for word in x.split() if word not in stop_words))
```

```python
df.head()
```

```python
pd.Series(" ".join(df['stopwords']).split()).value_counts()[:30]
```

```python
other_stop_words = ['one','came','even']
```

```python
df['cleanreviews'] = df['stopwords'].apply(lambda x: " ".join(word for word in x.split() if word not in other_stop_words))
```

```python
#df.head()
```

```python
pd.Series(" ".join(df['cleanreviews']).split()).value_counts()[:30]
```

```python
df['cleanreviews_word_count'] = df['stopwords'].apply(lambda x: len([word for word in x.split() if word not in other_stop_words]))
```

```python
df['clean_rate'] =  df['cleanreviews_word_count'] / df['word_count']
```

```python
df.head()
```

```python
#Lemmatization
```

```python
#import textblob
from textblob import Word
```

```python
df['lemmatized'] = df['cleanreviews'].apply(lambda x: " ".join(Word(word).lemmatize() for word in x.split()))
```

```python
#df.head()
```

```python
#(Markdown section) Sentiment Analysis
```

```python
from textblob import TextBlob
```

```python
df['polarity'] = df['lemmatized'].apply(lambda x: TextBlob(x).sentiment[0])
```

```python
df['subjectivity'] = df['lemmatized'].apply(lambda x: TextBlob(x).sentiment[1])
```

```python
df.head()
```

```python
df.describe()
```

```python
df.drop(['lowercase', 'punctuation', 'stopwords','cleanreviews', 'lemmatized'], axis=1, inplace=True)
```

```python
df.sort_values(by='polarity')
```
